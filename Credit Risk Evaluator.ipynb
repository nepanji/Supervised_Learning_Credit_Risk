```python
# Import Dependencies
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Load the data
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
# View train data
train_df
```

```python
# Create a training set from the 2019 loans csv
_2019_loans = train_df.drop(columns=['Unnamed: 0','index','loan_status'], axis=1)
_2019_loans
```

```python
_2019_loans.info()
```

```python
# Convert categorical data to numeric and separate target feature for training data
# One-hot encoding the entire dataframe
X_TR_Dummies=pd.get_dummies(_2019_loans)
print(X_TR_Dummies.columns)
X_TR_Dummies
```

```python
# Create a test set from the 2019 loans csv
y_tr_dummies = train_df['loan_status']
y_tr_dummies.unique()
```

```python
# Convert categorical data to numeric and separate target feature for testing data
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder

# Converting output labels to 0 and 1
y_tr_dummies = LabelEncoder().fit_transform(y_tr_dummies)
y_tr_dummies
```

```python
# View test info
test_df
```

```python
# Create a testing set from the 2020 loans csv
_2020_loans = test_df.drop(columns=['Unnamed: 0','index','loan_status'], axis=1)
_2020_loans
```

```python
_2020_loans.info()
```

```python
# Convert categorical data to numeric and separate target feature for training data
# One-hot encoding the entire dataframe
X_TE_Dummies=pd.get_dummies(_2020_loans)
print(X_TE_Dummies.columns)
X_TE_Dummies
```

```python
# Create a test set from the 2020Q1 loans csv
y_te_dummies = test_df['loan_status']
y_te_dummies.unique()
```

```python
# Convert categorical data to numeric and separate target feature for testing data

# Converting output labels to 0 and 1
y_te_dummies = LabelEncoder().fit_transform(y_te_dummies)
y_te_dummies
```

```python
test_df['debt_settlement_flag'].unique()
```

```python
# add missing dummy variables to testing set
# Add 2019 category to 2020 test set
add_data = X_TR_Dummies["debt_settlement_flag_Y"]
X_TE_Dummies = X_TE_Dummies.join(add_data)
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model
```

```python
# Fit the training model using the training data
model.fit(X_TR_Dummies, y_tr_dummies)
```

```python
# Print the training and test scores
print(f"Training Score: {model.score(X_TR_Dummies, y_tr_dummies)}")
print(f"Testing Score: {model.score(X_TE_Dummies, y_te_dummies)}")
```

```python
# Create Confusion Matrix of the predicted vs the actual outcomes
from sklearn.metrics import confusion_matrix

y_true = y_te_dummies
y_pred = model.predict(X_TE_Dummies)
confusion_matrix(y_true, y_pred)
```

```python
# Create the Classification Report and compare
from sklearn.metrics import classification_report

print(classification_report(y_true, y_pred))
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_TR_Dummies, y_tr_dummies)
```

```python
print(f'Training Score: {clf.score(X_TR_Dummies, y_tr_dummies)}')
print(f'Testing Score: {clf.score(X_TE_Dummies, y_te_dummies)}')
```

```python
# Create Confusion Matrix of the predicted vs the actual outcomes
from sklearn.metrics import confusion_matrix

y_true = y_te_dummies
y_pred = clf.predict(X_TE_Dummies)
confusion_matrix(y_true, y_pred)
```

```python
# Create the Classification Report and compare
from sklearn.metrics import classification_report

print(classification_report(y_true, y_pred))
```
